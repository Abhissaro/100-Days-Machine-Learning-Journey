```python
num_class = len(train.iloc[:,0].unique())
```

```python
x_train = train.iloc[:,1:]
y_train = train.iloc[:,0]

x_test = test.iloc[:,1:]
y_test = test.iloc[:,0]

from sklearn.preprocessing import scale
x_train = scale(x_train)
x_test=scale(x_test)

#print
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import lightgbm as lgb

lgtrain = lgb.Dataset(x_train, y_train)
lgval = lgb.Dataset(x_test, y_test)

params = {
          "objective" : "multiclass",           
          "max_depth": -1,
           "num_class":num_class,
          "learning_rate" : 0.0001,                 
          "verbosity" : 1 }


model = lgb.train(params, lgtrain, 500, valid_sets=[lgtrain, lgval], early_stopping_rounds=750, verbose_eval=200)
```

```python
# predict results
results = model.predict(x_test)

# select the index's with the maximum probability
results = np.argmax(results,axis = 1)

results = pd.Series(results,name="Label")
```
